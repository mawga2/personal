# **New feature(Illness-related)**

```python
df_with_disease_with_illness = merged_df_with_disease.copy()
df_illness = pd.read_csv(open("/content/drive/MyDrive/DESN9002/dataset_with_illness_classification.csv"))
df_with_disease_with_illness
```

```python
date=df_illness["created_at"]
counter = 0
for i in date:
    date[counter] = str(i)[0:10]
    counter += 1
df_illness["created_at"]=date
df_illness = df_illness.rename(columns = {'created_at':'Date'})
df_illness
```

```python
df_severity = df_illness [["illness"]]
df_severity.dropna()
df_severity
```

```python
OH_encodedd_df_severity = pd.get_dummies(df_severity)
OH_encodedd_df_severity["Date"] = df_illness["Date"]
OH_encodedd_df_severity
```

```python
OH_encodedd_df_severity = OH_encodedd_df_severity.groupby(['Date']).mean()
OH_encodedd_df_severity.sort_values('Date', ascending=True)
OH_encodedd_df_severity = OH_encodedd_df_severity.reset_index()
OH_encodedd_df_severity
```

```python
df_7days = OH_encodedd_df_severity.copy()
for i in df_7days:
    if (i == "illness_light illness") or (i == "illness_no illness") or (i == "illness_severe illness"):
      for j in range(6,376):
          df_7days[i][j] = (OH_encodedd_df_severity[i][j] + OH_encodedd_df_severity[i][j-1] + OH_encodedd_df_severity[i][j-2] + OH_encodedd_df_severity[i][j-3] + OH_encodedd_df_severity[i][j-4] + OH_encodedd_df_severity[i][j-5] + OH_encodedd_df_severity[i][j-6])/7
OH_encodedd_df_severity = df_7days.copy()
OH_encodedd_df_severity
```

```python
merged_df_with_DI = df_with_disease_with_illness.merge(OH_encodedd_df_severity, left_on = "Date", right_on = "Date")
merged_df_with_DI
```
